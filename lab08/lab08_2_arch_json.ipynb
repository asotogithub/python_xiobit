# Archivos JSON en Python

 Un Archivo json es un archivo de texto con extencion .json. Java Script Object Notation

Es un formato para intercambiar informacion. **JSON**

[Link de Definicion](https://en.wikipedia.org/wiki/JSON)

## Escritura de Archivos JSON, (Se llama Serealizacion)

```python
persona = {
    "nombre":"Juan",
    "apellido":"Perez",
    "edad":27,
    "es_casado":True,
    "direccion":{"calle":"C. Buenos Aires",
                "ciudad":"Cochabamba",
                "latitude":-16.53535,
                "longitude":37.41323
                },
    "notas":[41,
             56,
             47]
}
```

```python
type(persona)
```

```python
persona["edad"]
```

```python
persona['direccion']['latitude']
```

```python
sum(persona['notas'])
```

```python
# Importar la librera de json
```

```python
import json
```

```python
with open('alumno_perez.json', mode='w') as perez:
    json.dump(persona, perez)
```

## Lectura de Archivo JSON

```python
with open('alumno_perez.json', mode='r') as alumno:
    estudiante = json.load(alumno)
    
```

```python
estudiante
```

```python
type(estudiante)
```

## Consultar la informacion de un una pagina

```python
with open('recursos/all_day.geojson', mode = 'r') as terremotos:
    terremotos_dias = json.load(terremotos)
```

```python
type(terremotos_dias)
```

```python
terremotos_dias
```

```python
terremotos_lista = terremotos_dias["features"]
```

```python
terremotos_lista
```

```python
terremotos_lista[0]
```

```python
lista_terremotos[0]['properties']['mag']
```
